```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.set_printoptions(precision=3, suppress=True)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.compat.v2.keras.layers.experimental import preprocessing
```

```python
column_names = ["ratio",        # label is the output predicted
                "maximum",      # S+PGW maximum simultaneously active bearers (number)s
                "success",      # S+PGW successful bearer creations (times)
                "uplink",       # S1-U uplink user traffic in KB (kB)
                "downlink"]     # S1-U downlink user traffic in KB (kB)
```

```python
raw_dataset = pd.read_csv("final_ds.csv")
```

```python
dataset = raw_dataset.copy()
print(len(dataset["ratio"].unique().tolist()))
```

```python
#dataset
```

```python
dataset = pd.get_dummies(dataset, prefix='', prefix_sep='')
print(dataset.tail())
```

```python
# split dataset into train and test set
train = dataset.sample(frac=0.8, random_state=0)
test = dataset.drop(train.index)
```

```python
# inspectation of data sns.set(rc={'figure.figsize':(20,10)})
sns.pairplot(train[['ratio', "maximum", "success", "uplink", "downlink"]], diag_kind='kde')
plt.show()
```

```python
#train
```

```python
train.describe().transpose()
```

```python
train_features = train.copy()
test_features = test.copy()

train_labels = train_features.pop('ratio')
test_labels = test_features.pop('ratio')
```

```python
train.describe().transpose()[['mean', 'std']]
```

```python
normalizer = preprocessing.Normalization()
```

```python
normalizer.adapt(np.array(train_features))
```

```python
print(normalizer.mean.numpy())
```

```python
first = np.array(train_features[:1])
with np.printoptions(precision=2, suppress=True):
    print("First row data:", first)
    print()
    print("Normalized data:", normalizer(first).numpy())
```

```python
linear_model = tf.keras.Sequential([
    normalizer,
    layers.Dense(units=1)
])
#linear_model.summary()
```

```python
linear_model.predict(np.array(train_features[:10]))
```

```python
linear_model.layers[1].kernel
```

```python
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
history = linear_model.fit(
    train_features, train_labels, 
    epochs=100,
    # suppress logging
    verbose=0,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)
```

```python
plot_loss(history)
```

```python
def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.ylim([8, 80])
    plt.xlabel('Epoch')
    plt.ylabel('Error [ratio]')
    plt.legend()
    plt.grid(True)
```

```python
test_results = {}
```

```python
test_results['linear_model'] = linear_model.evaluate(
    test_features, test_labels, verbose=0)
```

```python
def build_and_compile_model(norm):
    model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

    model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
    return model
```

```python
### ======================================================
```

```python
up = np.array(train_features['uplink'])
up_normalizer = preprocessing.Normalization(input_shape=[1,])
up_normalizer.adapt(up)

up_model = tf.keras.Sequential([
    up_normalizer,
    layers.Dense(units=1)
])

up_model.summary()
```
